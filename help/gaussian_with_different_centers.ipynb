```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x = np.arange(0, 10, 0.1)
g_0 = norm(3)
g_1 = norm(5)
g_2 = norm(7)
```

```python
plt.plot(x, g_0.pdf(x))
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, sharey=True, figsize=(15, 3))
axs[0].plot(x, g_0.pdf(x))
axs[1].plot(x, g_1.pdf(x))
axs[2].plot(x, g_2.pdf(x))
```

```python
summed = g_0.pdf(x) + g_1.pdf(x) + g_2.pdf(x)
plt.plot(x, summed)
```
